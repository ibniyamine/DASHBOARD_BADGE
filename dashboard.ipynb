```python
df2 = pd.read_excel('resumer.xlsx')
df2.head()
```

```python
df2 = df2[["TYPE DE BADGES", "NB_EXACT"]]
df2.head()
```

```python
df2.rename(columns={"TYPE DE BADGES": "type", "NB_EXACT": "nb"}, inplace=True)
df2.head()
```

```python
df.rename(columns={"nb impression": "nb"}, inplace=True)
df.head()
```

```python
types_manquants = df2[~df2["type"].isin(df["type"])]
print(types_manquants)
```

```python
df["type"].unique()
```

```python
types_manquants = df[~df["type"].isin(df2["type"])]
types_manquants
```

```python
df = df.merge(df2[["type", "nb"]], on="type", how="left", suffixes=("", "_df2"))

df["nb"] = df.groupby("type")["nb_df2"].transform(
    lambda x: [x.iloc[0]] + [0]*(len(x)-1)
)

df = df.drop(columns=["nb_df2"])
df.head()
```

```python
types_absents = set(df2["type"]) - set(df["type"])
lignes_a_ajouter = df2[df2["type"].isin(types_absents)].copy()
lignes_a_ajouter["structure"] = lignes_a_ajouter["type"]
lignes_a_ajouter["date"] = None   # ou pd.NaT si tu préfères
lignes_a_ajouter = lignes_a_ajouter[["nb", "structure", "type", "date"]]
df = pd.concat([df, lignes_a_ajouter], ignore_index=True)
df.head()
```

```python
types_manquants = df2[~df2["type"].isin(df["type"])]
print(types_manquants)
```

```python
df["type"].unique()
```

```python
df["type"].value_counts()   
```

```python
groupe = df.groupby("type")["nb"].sum().sort_values(ascending=False)
groupe
```

```python
nb_seter = df.loc[df["type"] == "SETER", "nb"].sum()
df.loc[df["type"] == "COMMISSION ORGANISATION", "nb"] += nb_seter
df = df[df["type"] != "SETER"]
```

```python
df["type"].unique()
```

```python
groupe = df.groupby("type")["nb"].sum().sort_values(ascending=False)
groupe
```

```python
nb_chauffeurs = df["type"].value_counts().get("CHAUFFEUR", 0)
nb_chauffeurs
```

```python
commission_mobilisation = df["type"].value_counts().get("COMMISSION MOBILISATION", 0)
commission_mobilisation
```

```python
partenaire_fsf = df["type"].value_counts().get("PARTENAIRE FSF", 0)
partenaire_fsf
```

```python


# Dictionnaire des types et valeurs à attribuer
rules = {
    "CHAUFFEUR": 20,
    "COMMISSION MOBILISATION": 8,
    "PARTENAIRE FSF": 1
}

# Pour chaque type, on trouve la première ligne et on met la valeur
for t, val in rules.items():
    idx = df.index[df["type"] == t]
    if len(idx) > 0:
        df.loc[idx[0], "nb"] = val
```

```python
groupe = df.groupby("type")["nb"].sum().sort_values(ascending=False)
groupe
```

```python
df_sum = df["nb"].sum()
df_sum
```

```python
mask = (df["type"] == "COMMISSION ORGANISATION") & (df["nb"] != 0)
df.loc[mask, "nb"] = 109
df_sum = df["nb"].sum()
df_sum
```

```python
# On met d'abord toutes les lignes de ce type à 0
df.loc[df["type"] == "COMMISSION ORGANISATION", "nb"] = 0

# On récupère l'index de la première occurrence
idx = df.index[df["type"] == "COMMISSION ORGANISATION"]

# Si au moins une ligne existe, on met 109 à la première
if len(idx) > 0:
    df.loc[idx[0], "nb"] = 109
df_sum = df["nb"].sum()
df_sum
```
